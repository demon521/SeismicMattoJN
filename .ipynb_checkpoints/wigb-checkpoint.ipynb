```python
def wigb(a,scal,amx):
    [nz,nx]=a.shape
    trmx= np.max(abs(a))
# take the average as dx
    x=np.arange(0,nx)
    z=np.arange(0,nz)
    dx1 = abs(x[2:nx]-x[1:nx-1])
    dx = np.median(dx1)
    
    dz=z[1]-z[0]
    xmx=np.max(np.max(a))
    xmn=np.min(np.min(a))
    a = a * dx /amx;
    print(' PlotWig: data range',xmn,xmx,'plotted max',amx)
    #% set display range 

    x1=min(x)-2.0*dx 
    x2=max(x)+2.0*dx
    z1=min(z)-dz
    z2=max(z)+dz
 
    z=z.T; #% input as row vector
    zstart=z[1]
    zend=z[nz-1]
    
    for i in range(nx):
        if trmx[i] != 0:
            tr=a[:,i]
            s = np.sign(tr)
            i1= np.array(np.where( s[1:nz-1] != s[2:nz]))+1
            npos = np.size(i1);

            zadd = i1 +1+ tr[i1] / (tr[i1] - tr[i1+1])
            aadd = np.zeros(np.size(zadd))

            zpos =np.array(np.where(tr >0))+1
            zz=np.sort([np.hstack((zpos,zadd))])
            zz=zz.reshape(1,np.size(zz))
            iz=np.argsort([np.hstack((zpos,zadd))])
            aa = np.append(tr[zpos-1].T,aadd)
            aa = aa[iz]
            aa= aa.reshape(1,np.size(aa))
            if tr[0]>0:
                a0=0
                z0=1.00
            else:
                a0=0
                z0=zadd[0]
            if tr[nz-1]>0:
                a1=0 
                z1=nz 
            else:
                a1=0
                z1=max(zadd)
            zz=np.insert(zz, 0, z0)
            zz=np.append(zz,z1)
            zz=np.append(zz,z0)
            aa=np.insert(aa, 0, a0)
            aa=np.append(aa,a1)
            aa=np.append(aa,a0)

            zzz = zstart + zz*dz -dz;
            aaa = aa+x[i]
            tempp=[]
            for j in range(np.size(zzz)):
                tempp=np.append(tempp,[aaa[j],zzz[j]])
            tempp=tempp.reshape(np.size(zzz),2)
        
        
        
        fig=plt.figure(figsize=(10, 10))
        pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]], color = 'g', alpha = 0.5 )
        ax1.add_patch(pgon)
        plt.show()
```
